```python
def make_sphere(n_points, mean_radius=1.0, surface=True):
    delta_radius = 1.0
    phi = random.random(n_points)*2.0*pi
    costheta = 2.0*(random.random(n_points)-0.5)
    theta = arccos(costheta)
    if(surface):
        radius = mean_radius * ones(n_points)
    else:
        radius = mean_radius * random.random(n_points)**(1.0/3.0)
        
    x = radius * sin(theta) * sin(phi)
    y = radius * sin(theta) * cos(phi)
    z = radius * cos(theta)
    if surface==True:
        mean_r = mean_radius
    else:
        mean_r = mean_radius * (1.0/n_points)**(1.0/3.0)
    return [x,y,z, mean_r]
```

```python
def force(x,y,z, mass=1.0):
    radius = x**2 + y**2 + z**2
    force_x = mass*x/(radius)**(3.0/2.0)
    force_y = mass*y/(radius)**(3.0/2.0)
    force_z = mass*z/(radius)**(3.0/2.0)
    total_force_x = sum(force_x)
    total_force_y = sum(force_y)
    total_force_z = sum(force_z)
    norm_total_force = sqrt(total_force_x**2 + total_force_y**2 + total_force_z**2)
    return norm_total_force, total_force_x, total_force_y, total_force_z
```

```python
n_part = 1.0E4*ones(10000)
f_total = empty((0))
mean_radius = 10.0
for n in n_part:
    x,y,z, mean_r = make_sphere(n, mean_radius=mean_radius, surface=False)
    total_force,f_x,f_y,f_z = force(x,y,z)
    total_force = total_force*mean_r**2
    f_total = append(f_total, total_force)
```

Here I do some testing based on some large N-body simulation

```python
def n_body_force(sim_x, sim_y, sim_z, sim_mass, center_fiducial=500.0, sigma_center=300.0, radius_sphere=300.0,
                inner_radius=300.0, center_x=-1.0, center_y=-1.0, center_z=-1.0):
    box_size = 1000.0
    #select a random position in the box corresponding to some halo
    if((center_x<0.0) or (center_y<0.0) or (center_z<0.0)):
#        center_x = random.normal(center_fiducial, sigma_center)
#        center_y = random.normal(center_fiducial, sigma_center)
#        center_z = random.normal(center_fiducial, sigma_center)
        random_id  = random.randint(0,size(sim_x))
        center_x = sim_x[random_id]
        center_y = sim_y[random_id]
        center_z = sim_z[random_id]
     
    #find the halos inside the box
    x_tmp = (pos_x - center_x)%box_size
    y_tmp = (pos_y - center_y)%box_size
    z_tmp = (pos_z - center_z)%box_size
    
    rad_tmp = sqrt(x_tmp**2 + y_tmp**2 + z_tmp**2)
    
    #these are the positions inside a shell
    inside = where((rad_tmp< radius_sphere)&(rad_tmp>inner_radius)&(rad_tmp>0.001))
    x_new = x_tmp[inside]
    y_new = y_tmp[inside]
    z_new = z_tmp[inside]
    mass_new = mass[inside]
    
    inner = where((rad_tmp<=inner_radius)&(rad_tmp>0.001))
    x_inner = x_tmp[inner]
    y_inner = y_tmp[inner]
    z_inner = z_tmp[inner]
    mass_inner = mass[inner]

    #compute the force at the center
    total_force,f_x,f_y,f_z = force(x_new, y_new, z_new, mass=mass_new)
    total_force_inner,f_x_in,f_y_in,f_z_in = force(x_inner, y_inner, z_inner, mass=mass_inner)

    
    #normalize it to the force of the average mass located at the sphere's mean separation
    mean_mass = average(mass_new)
    mean_separation = ((radius_sphere**3-inner_radius**3)/size(mass_new))**(1.0/3.0)
    mean_force = mean_mass/(mean_separation**2)
   
    return total_force, mean_force, total_force_inner, mean_mass, mean_separation
```

```python
min_vel = 200
max_vel = 1000
filename = "../data/MD_full_halos_%d-%dkms_z0.dat"%(min_vel, max_vel)
BDMW_data = loadtxt(filename, skiprows=37, delimiter=',')
```

```python
pos_x = BDMW_data[:,4]
pos_y = BDMW_data[:,5]
pos_z = BDMW_data[:,6]
mass = BDMW_data[:,12]
```

Traces a line through the simulation

```python
n_samples = 1000
center_x = linspace(0.0,1000.0,n_samples)
center_y = ones(n_samples) * 100.0
center_z = ones(n_samples) * 350.0
total_force_line = empty((0))
inner_force_line = empty((0))
mean_force_line = empty((0))
distance_line = empty((0))
r_in = 50.0
r_out = 500.0
for i in range(n_samples):
    tmp_total, tmp_mean, tmp_inner, tmp_mass, tmp_separation = \
        n_body_force(pos_x, pos_y, pos_z, mass, radius_sphere=r_out, inner_radius=r_in,
                     center_x = center_x[i], center_y = center_y[i], center_z=center_z[i])
    if(tmp_inner>0):
        total_force_line = append(total_force_line, tmp_total)
        inner_force_line = append(inner_force_line, tmp_inner)
        mean_force_line  = append(mean_force_line, tmp_mean)
        distance_line  = append(distance_line, center_x[i])
print tmp_mass, tmp_separation
```

```python
fig = figure(1, figsize=(9.5,6.5))
ax = axes()
ax.set_yscale('log')

ax.set_xlabel("$\mathrm{distance\ [Mpc}/h]$",fontsize=25)
ax.set_ylabel("$\mathrm{\log_{10} F_{out} / F_{in}}$",fontsize=25)

ticklabels_x = ax.get_xticklabels()
ticklabels_y = ax.get_yticklabels()

for label_x in ticklabels_x:
    label_x.set_fontsize(18)
    label_x.set_family('serif')
for label_y in ticklabels_y:
    label_y.set_fontsize(18)
    label_y.set_family('serif')
    
label="$\mathrm{R_{in}=%d\ Mpc}/h$\n$\mathrm{R_{out}=%d\ Mpc}/h$"%(r_in, r_out)
plt.text(200, 0.5, label, ha='left', fontsize=25)

plot(distance_line, total_force_line/inner_force_line, linewidth=1)
print tmp_mass, tmp_separation

ax.set_xlim([0,1000.0])

#legend(numpoints=1,scatterpoints=1, loc=2)

filename='trace_nbody_%d_%d'%(min_vel, max_vel)
savefig(filename + '.eps',format = 'eps', transparent=True)
savefig(filename + '.pdf',format = 'pdf', transparent=True)
savefig(filename + '.png',format = 'png', transparent=True)
```

I was reading  "3D Velocity and Density Reconstructions of the Local Universe with Cosmicflows-1" [http://arxiv.org/abs/1109.3856]. There they quote a velocity for the LG of 593 km/s. This is dissected in two parts. One is local due to the matter within 30Mpc/h and the other is tidal due to the matter outside that region. The local component is 304km/s and the tidal is 382kms/. This makes sense in the terms we have presented here. The tidal force (which can go higher than 300Mpc/h) is comparable (and a bit larger) to the force caused by a local group of matter.
