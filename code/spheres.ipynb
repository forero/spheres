```python
def make_sphere(n_points, mean_radius=1.0, surface=True):
    delta_radius = 1.0
    phi = random.random(n_points)*2.0*pi
    costheta = 2.0*(random.random(n_points)-0.5)
    theta = arccos(costheta)
    if(surface):
        radius = mean_radius * ones(n_points)
    else:
        radius = mean_radius * random.random(n_points)**(1.0/3.0)
        
#    radius = random.normal(mean_radius, delta_radius, n_points)
#    radius = (mean_radius * random.random(n_points))+ delta_radius
    x = radius * sin(theta) * sin(phi)
    y = radius * sin(theta) * cos(phi)
    z = radius * cos(theta)
    if surface==True:
        mean_r = mean_radius
    else:
        mean_r = mean_radius * (1.0/n_points)**(1.0/3.0)
    return [x,y,z, mean_r]
```

```python
def force(x,y,z, mass=1.0):
    force_x = mass*x/(x**2 + y**2 + z**2)**(3.0/2.0)
    force_y = mass*y/(x**2 + y**2 + z**2)**(3.0/2.0)
    force_z = mass*z/(x**2 + y**2 + z**2)**(3.0/2.0)
    total_force_x = sum(force_x)
    total_force_y = sum(force_y)
    total_force_z = sum(force_z)
    norm_total_force = sqrt(total_force_x**2 + total_force_y**2 + total_force_z**2)
    return norm_total_force, total_force_x, total_force_y, total_force_z
```

```python
n_part = [1.0,10.0,1.0E2,1.0E3,1.0E4, 1.0E5, 1.0E6, 1.0E7]
n_part = 1.0E3*ones(10000)
f_total = empty((0))
mean_radius = 10.0
for n in n_part:
    x,y,z, mean_r = make_sphere(n, mean_radius=mean_radius, surface=False)
    total_force,f_x,f_y,f_z = force(x,y,z)
    total_force = total_force*mean_r**2
    f_total = append(f_total, total_force)
#    print f_total
#    print total_force
```

```python
filename = "../../SparseFilaments/data/input/MD_full_halos_900-1000kms_z0.dat"
BDMW_data = loadtxt(filename, skiprows=37, delimiter=',')
```

```python
pos_x = BDMW_data[:,4]
pos_y = BDMW_data[:,5]
pos_z = BDMW_data[:,6]
mass = BDMW_data[:,12]
```

```python
def n_body_force(sim_x, sim_y, sim_z, sim_mass, center_fiducial=500.0, sigma_center=20.0, radius_sphere=300.0):
   
    #select a random position in the box.
    center_x = random.normal(center_fiducial, sigma_center)
    center_y = random.normal(center_fiducial, sigma_center)
    center_z = random.normal(center_fiducial, sigma_center)
    
    #check that a sphere centered at that point will be containe inside the box
    if(((center_x-radius_sphere)<0.0) | ((center_y-radius_sphere)<0.0)| ((center_z-radius_sphere)<0.0)):
        print 'warning'
    if(((center_x+radius_sphere)>1000.0) | ((center_y-radius_sphere)>1000.0)| ((center_z-radius_sphere)>1000.0)):
        print 'warning'
        
    #find the halos inside the box
    x_tmp = pos_x - center_x
    y_tmp = pos_y - center_y
    z_tmp = pos_z - center_z
    rad_tmp = sqrt(x_tmp**2 + y_tmp**2 + z_tmp**2)
    inside = where((rad_tmp< radius_sphere)&(rad_tmp>sigma_center))
    
    x_new = x_tmp[inside]
    y_new = y_tmp[inside]
    z_new = z_tmp[inside]
    mass_new = mass[inside]

    #compute the force at the center
    total_force,f_x,f_y,f_z = force(x_new, y_new, z_new, mass=mass_new)
    
    #normalize it to the force of the average mass located at the sphere's surface
    mean_mass = average(mass_new)
    mean_separation = ((radius_sphere-sigma_center)**3/size(mass_new))**(1.0/3.0)
    mean_force = amax(mass_new)/(mean_separation**2)
    return total_force/mean_force
```

```python
average_force = empty((0))

for i in range(1000):
    tmp_average = n_body_force(pos_x, pos_y, pos_z, mass, radius_sphere=400.0)
    average_force = append(average_force, tmp_average)
```

```python
n, bins, patches = hist(average_force, 20, normed=0, color = 'silver')
```
