```python
n_points = 100000
min_l = 0.0
max_l = 10.0
delta_l = (max_l - min_l)/n_points
```

```python
x_pos = linspace(min_l, max_l, n_points)
```

```python
n, bins, patches = hist(x_pos, 100, normed=1, color = 'silver')
```

```python
k_x = 0.5
x_new = x_pos + 0.1*delta_l*sin(2.0*pi*k_x*x_pos)
```

```python
n, bins, patches = hist(x_new, 50, normed=1, color = 'silver')
```

```python
plot(x_pos-x_new)
```

```python
x_0 = 4.96
mass = 1.0/n_points
force_x = mass*x_pos/(x_0 - x_pos)**(3.0)
force_x_new = mass*x_new/(x_0 - x_new)**(3.0)
```

```python
plot(log10(abs(force_x - force_x_new)))
```
